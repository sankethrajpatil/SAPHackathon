```python
import pandas as pd 
#loading the dataset
df = pd.read_excel('Burnout Rate - Before modelling.xlsx')
df.head()
```

```python
data = df[['burnout', 'Employee_burnout']]
data
```

```python
#Since the last column is in words, it has to be converted into a suitable scale
data['scale']=0
data.loc[data['Employee_burnout']=='No burnout','scale']= 0
data.loc[data['Employee_burnout']=='Alarming burnout','scale']= 1
data.loc[data['Employee_burnout']=='Moderate burnout','scale']= 2
data.loc[data['Employee_burnout']=='Total burnout','scale']= 4
data
```

```python
#loading the columns in x and y variables
x = data['burnout']
#Reshaping the x columns as a 2D array is expected here
x = x.values.reshape(-1,1)
y = data['scale']
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[40.221]])
```
