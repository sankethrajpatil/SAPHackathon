```python
#Importing the libraries
import pandas as pd
import random
import numpy as np
```

```python
#Loading the dataset in a variable
df = pd.read_excel('Burnout Rate.xlsx')
#Reading the first 5 columns of the dataset
df.head()
```

```python
df.info()
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df.head()
```

```python
#Function To Print random numbers in project ID
list1 = [] #EmptyList
def proid():
    l1 = random.randint(1, 15)
    list1.append(l1)
```

```python
#Function To Print random numbers in Missed deadlines
list2 = [] #EmptyList
def mdead():
    l2 = random.randint(0, 15)
    list2.append(l2)  
```

```python
#Function To Print random numbers in number of unplanned meetings
list3 = [] #EmptyList
def unplan():
    l3 = random.randint(3, 15)
    list3.append(l3)
```

```python
#Function To Print random numbers in Number of meetings organized
list4 = [] #EmptyList
def meorg():
    l4 = random.randint(0, 8)
    list4.append(l4)
```

```python
#Function To Print random numbers in Number of meetings participated
list5 = [] #EmptyList
def mepart():
    l5 = random.uniform(0, 15)
    list5.append(l5)
```

```python
#Function To Print random numbers in avg meeting participated
list6 = [] #EmptyList
def avgme():
    l6 = random.uniform(0, 6)
    list6.append(l6)
```

```python
#Function To Print random numbers in office laptop
list7 = [] #EmptyList
def olap():
    l7 = random.uniform(8, 16)
    list7.append(l7)
```

```python
for i in range(0,200):
    proid()
    mdead()
    unplan()
    meorg()
    mepart()
    avgme()
    olap()

#Calling the functions so that the data is collected in the respective lists
```

```python
#Assigning each empty column with the list that we have made
df['Project ID'] = list1
df['Missed deadlines'] = list2
df['Number of unplanned meetings'] = list3
df['Number of meetings organized'] = list4
df['Number of meetings participated'] = list5
df['Average meeting time per day/week'] = list6
df['On an office laptop'] = list7
```

```python
#Checking the dataframe
df.head()
```

```python
#checking which dataset has empty values
df.isna().sum()
```

```python
#converting the columns into the right the data type
df['Time/Duration'] = df['Time/Duration'].astype(float)
df['General leave patterns'] = df['General leave patterns'].astype(float)
df['Sick leave(s)'] = df['Sick leave(s)'].astype(float)
df['On an office laptop'] = df['On an office laptop'].astype(float)
```

```python
#Filling the empty values with mean value
df['Performance'] = df['Performance'].fillna(df['Performance'].mean())
df['Time/Duration'] = df['Time/Duration'].fillna(df['Time/Duration'].mean())
df['General leave patterns'] = df['General leave patterns'].fillna(df['General leave patterns'].mean())
df['Sick leave(s)'] = df['Sick leave(s)'].fillna(df['Sick leave(s)'].mean())
```

```python
#checking final value
df.info()
```

```python
df.to_excel('Burnout rate - After Cleanse.xlsx')
```
